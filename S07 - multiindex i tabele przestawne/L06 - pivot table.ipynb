```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv(r"..\course-files\marathon_results_2016.csv", index_col='Bib',
                usecols=['Bib', '40K', 'Half', 'Pace', 'Age', 'M/F', 'Country', 'State', 'City'])
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
```

```python
df.TotalSeconds = df.TotalSeconds.replace(0, np.NaN)
df.HalfSeconds = df.HalfSeconds.replace(0, np.NaN)
```

```python
df.pivot_table(index='Age', columns="M/F", values='TotalSeconds').head()
```

```python
df.pivot_table(index='Age', columns="M/F", values='TotalSeconds').sort_values('M')
```

```python
df.pivot_table(index='Age', columns="M/F", values='TotalSeconds').sort_values('F')
```

```python
df.pivot_table(index='Age', columns="M/F", values=['TotalSeconds', 'HalfSeconds']).head()
```

```python
pivotdf = df.pivot_table(index='Age', columns="M/F", values=['TotalSeconds', 'HalfSeconds'])
```

```python
pivotdf.swaplevel(axis=1).sort_index(axis=1).head()
```

```python
df.pivot_table(index='Age', columns="M/F", values='TotalSeconds', aggfunc='count').head()
```

```python
df.pivot_table(index='Age', values='TotalSeconds', aggfunc='count').head()
```

```python
df.pivot_table(index='Country', values='40K', aggfunc='count').head()
```

```python
df.pivot_table(index='Country', values='40K', aggfunc='count').sort_values('40K', ascending=False).head()
```

```python
df.pivot_table(index=['Country', 'State', 'City'], values='TotalSeconds').sort_index().head(10)
```

```python
df.pivot_table(index=['Country', 'State', 'City'], values='TotalSeconds').sort_values('TotalSeconds', ascending=False).head(10)
```
