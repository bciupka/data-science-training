melt odpivotuje df. id_vars określa co ma być indexem, value_vars - co ma byc odpiwvotowane - zawężenie (kolumny na wiersze), value name - nazwa kolumny z wartościami, var name - nazwa nowo utworzonej kolumny z nagłówków kolumn

melt powinno się używać jeżeli dane pobrane do df są spivotowane. Jeżeli samemu najpierw wykonywało się operacje używać unstack.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../course-files/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values="TotalSeconds").head()
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.melt(id_vars='Age', value_name='TotalSeconds', var_name='Sex')
```

```python
df = pd.read_csv('../course-files/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"]).head()
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.melt(id_vars='Age', value_name='Time')
```
