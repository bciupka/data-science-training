**outer merga najlepiej używa się do identyfiacji, jakie dane pochodzą z którego zbioru**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
suppliers = pd.read_csv(r"..\course-files\northwind-mongo-master\suppliers_del_1.csv", usecols=['SupplierID', 'CompanyName', 'City'])
suppliers.head()
```

```python
products = pd.read_csv(r'../course-files/northwind-mongo-master/products_del_1.csv', usecols=['ProductID', 'ProductName', 'SupplierID'])
products.head()
```

```python
prod_and_sup = suppliers.merge(products, on='SupplierID')
prod_and_sup.head()
```

```python
print(len(products), len(suppliers), len(prod_and_sup))
```

```python
supDel = suppliers.merge(products, how='right',on='SupplierID', indicator=True)
supDel.head()
```

```python
supDel[supDel._merge == 'right_only']
```

```python
supUnUse = suppliers.merge(products, how='left', on='SupplierID', indicator=True)
```

```python
supUnUse[supUnUse._merge == 'left_only']
```
