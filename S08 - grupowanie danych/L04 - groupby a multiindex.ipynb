```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../course-files/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country',
'State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
country_city = df.groupby(['Country', 'City'])
```

```python
country_city.size()
```

```python
country_city[['Age', 'Half', '40K', 'TotalSeconds', 'HalfSeconds']].mean()
```

```python
country_city.count()
```

```python
country_city.get_group(('USA', 'San Francisco'))
```

```python
sex_age = df.groupby(['M/F', 'Age'])
```

```python
sex_age.size().head()
```

```python
sex_age[['Age', 'Half', '40K', 'TotalSeconds', 'HalfSeconds']].mean().head()
```

```python
sex_age.get_group(('M', 25))
```
