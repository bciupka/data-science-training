```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
fortune = pd.read_csv(r'..\course-files\Fortune_500_2017.csv', usecols=['Rank', 'Title', 'Industry', 'Hqlocation', 'Sector'])
fortune.head()
```

```python
fortune.Sector = fortune.Sector.str.upper()
fortune.head()
```

```python
fortune[fortune.Industry.str.lower().str.contains('comp')].head()
```

```python
fortune[['City', 'State']] = fortune.Hqlocation.str.split(',', expand=True) #[[]] - LISTA KOLUMN
fortune.head()
```

```python
def BuildShortcut(row):
    temp = row.Industry.split(' ')
    temp2 = ''
    for i in temp:
        temp2 += i[0]
    return temp2
```

```python
fortune['IndustryShort'] = fortune.apply(BuildShortcut, axis=1).head()
```
